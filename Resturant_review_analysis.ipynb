```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
import string
string.punctuation
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def transform_text(text):
  text=text.lower()
  text=nltk.word_tokenize(text)

  y=[]
  for i in text:
    if i.isalnum():
      y.append(i)
  text=y[:]
  y.clear()

  for i in text:
    if i not in stopwords.words('english') and i not in string.punctuation:
      y.append(i)

  text=y[:]
  y.clear()

  for i in text:
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
df['Review'].apply(transform_text)
```

```python
df['transformed_text']=df['Review'].apply(transform_text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['transformed_text'],df['Liked'],test_size=0.3,random_state=0,shuffle=True)
```

# **Building Model Pipeline Tf-idf (Random Forest)**



```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier=Pipeline([('tfidf',TfidfVectorizer()),('clf',RandomForestClassifier(n_estimators=100))])
```

```python
rf_classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
rf_pred=rf_classifier.predict(x_test)
```

```python
#Confusion Matrix
confusion_matrix(y_test,rf_pred)
```

```python
#Accuracy Score
accuracy_score(y_test,rf_pred)
```

```python
#Classification Report
print(classification_report(y_test,rf_pred))
```
